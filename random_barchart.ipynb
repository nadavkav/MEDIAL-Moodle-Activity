# Random Bar Chart Generator
This notebook generates a bar chart with random values using Python's `random` and `matplotlib` libraries.

```python
import random
import matplotlib.pyplot as plt

# Categories and random values
categories = ['A', 'B', 'C', 'D', 'E']
values = [random.randint(1, 100) for _ in categories]

# Plot
plt.figure(figsize=(8, 5))
plt.bar(categories, values)
plt.title('Random Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y')
plt.show()
print('hello')
```
